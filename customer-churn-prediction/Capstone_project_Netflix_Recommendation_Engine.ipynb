```python
#pre filtered the dataset:
#prefilter means in our data it will contain unreliable data and unrated data
#it means it contain less rating movie and unrated users in customer_id
```

```python
#count the cust_id from rating
dataset_cust_summary=netflix_dataset.groupby('cust_id')['Rating'].agg(['count'])
dataset_cust_summary
```

```python
#create abenchmark for cust_id to filter the data here we will use 60%
cust_benchmark=round(dataset_cust_summary['count'].quantile(0.6),0)
cust_benchmark
```

```python
#60% of customer below the 36 rating and 40% of data have 36 and more
```

```python
drop_cust=dataset_cust_summary[dataset_cust_summary['count']<cust_benchmark].index
drop_cust
```

```python
len(drop_cust)
```

```python
#282042 user is remove from dataset
```

```python
#similarly we will do for movies
movie_summary=netflix_dataset.groupby('movie_id')['Rating'].agg(['count'])
movie_summary
```

```python
#create a bench mark for a movie
movie_benchmark=round(movie_summary['count'].quantile(0.6),0)
movie_benchmark
```

```python
#60% of movie below the 908 rating and 40% of movie above 908 and more
```

```python
#drop the movie below 908 rating
drop_movie=movie_summary[movie_summary['count']<movie_benchmark].index
drop_movie
```

```python
len(drop_movie)
```

```python
#2699 movie to be remove from datset
```

```python
#now chage in dataset by using drop data
netflix_dataset=netflix_dataset[~netflix_dataset['movie_id'].isin(drop_movie)]
netflix_dataset=netflix_dataset[~netflix_dataset['cust_id'].isin(drop_cust)]
```
